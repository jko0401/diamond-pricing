```python
import requests
import json
import time
import sys
import random 
import pandas as pd
```

```python
shapes = ['OV', 'CU']
# ['RD', 'PR','CU','MQ','OV','PS','RA','HS', 'EC','AS']
cuts = ['Good', 'Very%20Good', 'Ideal', 'Astor%20Ideal']
colors = ['K', 'J', 'I', 'H', 'G', 'F', 'E', 'D']
clarities = ['SI2', 'SI1', 'VS2', 'VS1', 'VVS2', 'VVS1', 'IF', 'FL']
diamonds_dict = {}

for shape in shapes:
    for cut in cuts:
        for color in colors:
            for clarity in clarities:
                source_url = "https://www.bluenile.com/api/public/diamond-search-grid/v2?startIndex=0&pageSize=1000&_=1591725873076&unlimitedPaging=false&sortDirection=desc&sortColumn=price&shape={}&minCut={}&maxCut={}&minColor={}&maxColor={}&minClarity={}&maxClarity={}&maxDateType=MANUFACTURING_REQUIRED&isQuickShip=false&hasVisualization=false&isFiltersExpanded=false&astorFilterActive=false&country=USA&language=en-us&currency=USD&productSet=BN".format(shape, cut, cut, color, color, clarity, clarity)
                print(source_url)

                while True:
                    try:
                        data = requests.get(source_url).json()
                        query = 'Querying: {}, {}, {}, {}'.format(shape, cut, color, clarity)
                        break
                    except Exception as e:
                        print(e)
                        wait = random.randrange(20);
                        print('Error! Waiting %s secs and re-trying...' % wait)
                        sys.stdout.flush()
                        time.sleep(wait)

                total_count = data['countRaw']
                print(query + '; Total Count: ' + str(total_count))
                if total_count > 1000:
                    total_count = 1000
                for i in range(total_count):
                    diamond_id = data['results'][i]['id']
                    diamonds_dict[diamond_id] = {'detailsPageUrl': '', 'carat': '', 'date': '', 'dateSet': '', 'price': '', 
                                                'strikethroughPrice': '', 'pricePerCarat': '', 'shapeName': '', 'clarity': '', 
                                                'color': '', 'culet': '', 'cut': '', 'depth': '', 'fluorescence': '', 
                                                'lxwRatio': '', 'polish': '', 'symmetry': '', 'table': '', 'measurements': ''}
                    for attribute in diamonds_dict[diamond_id].keys():
                        if attribute is 'cut' or attribute is 'measurements':
                            diamonds_dict[diamond_id][attribute] = data['results'][i][attribute][0]['label']
                        else:
                            diamonds_dict[diamond_id][attribute] = data['results'][i][attribute]
                
                print(str(len(diamonds_dict.keys())) + ' diamonds added.')
```

```python
diamonds_df = pd.DataFrame.from_dict(diamonds_dict, orient='index')
```

```python
diamonds_df.to_csv('bluenile_diamonds_ov-cu.csv')
```

```python
shapes = ['Round', 'Oval', 'Cushion']
# ['Princess', 'Pear', 'Emerald', 'Marquise', 'Asscher', 'Radiant', 'Heart']
cuts = ['Fair', 'Good', 'Very%20Good', 'Ideal', 'Super%20Ideal']
colors = ['J', 'I', 'H', 'G', 'F', 'E', 'D']
clarities = ['SI2', 'SI1', 'VS2', 'VS1', 'VVS2', 'VVS1', 'IF', 'FL']
brilliant_dict = {}

for shape in shapes:
    for cut in cuts:
        for color in colors:
            for clarity in clarities:
                brilliant_url = 'https://www.brilliantearth.com/loose-diamonds/list/?shapes={}&cuts={}&colors={}&clarities={}&polishes=Good%2CVery+Good%2CExcellent&symmetries=Good%2CVery+Good%2CExcellent&fluorescences=Very+Strong%2CStrong%2CMedium%2CFaint%2CNone&min_carat=0.25&max_carat=17.55&min_table=50.00&max_table=86.00&min_depth=46.20&max_depth=629.00&min_price=410&max_price=1312550&stock_number=&row=0&page=1&requestedDataSize=2000&order_by=price&order_method=asc&currency=%24&has_v360_video=&dedicated=&min_ratio=1.00&max_ratio=2.75&shipping_day=&MIN_PRICE=410&MAX_PRICE=1312550&MIN_CARAT=0.25&MAX_CARAT=17.55&MIN_TABLE=45&MAX_TABLE=86&MIN_DEPTH=46.2&MAX_DEPTH=629'.format(shape, cut, color, clarity)
                while True:
                    try:
                        data = requests.get(brilliant_url).json()
                        query = 'Querying: {}, {}, {}, {}'.format(shape, cut, color, clarity)
                        print(query)
                        break
                    except Exception as e:
                        print(e)
                        wait = random.randrange(20);
                        print('Error! Waiting %s secs and re-trying...' % wait)
                        sys.stdout.flush()
                        time.sleep(wait)
                diamonds_queried = len(data['diamonds'])
                print('Number of Diamonds Queried: ' + str(diamonds_queried))
                if diamonds_queried != 0:
                    for i in range(diamonds_queried):
                        diamond_id = data['diamonds'][i]['upc']
                        brilliant_dict[diamond_id] = {'carat': '', 'price': '', 
                                                    'shape': '', 'clarity': '', 'origin': '', 
                                                    'color': '', 'culet': '', 'cut': '', 'depth': '', 'fluorescence': '', 
                                                    'length_width_ratio': '', 'polish': '', 'symmetry': '', 'table': '', 'measurements': ''}
                        for attribute in brilliant_dict[diamond_id].keys():
                                brilliant_dict[diamond_id][attribute] = data['diamonds'][i][attribute]
                print(str(len(brilliant_dict.keys())) + ' diamonds added.')
```

```python
print(len(brilliant_dict))
```

```python
brilliant_df = pd.DataFrame.from_dict(brilliant_dict, orient='index')
```

```python
brilliant_df.reset_index(inplace=True, drop=True)
brilliant_df[['measurements', 'width', 'height']] = brilliant_df['measurements'].str.split('x', expand=True).astype(float)
brilliant_df.rename(columns={'index': 'id',
                             'measurements': 'length', 
                             'depth': 'depth%', 
                             'table': 'table%',
                             'shape': 'shapeName',
                             'length_width_ratio': 'lxwRatio'}, inplace=True)
brilliant_df['pricePerCarat'] = brilliant_df['price']/brilliant_df['carat']
```

```python
brilliant_df.to_csv('brilliant_diamonds_rd-ov-cu.csv', index=False)
```
